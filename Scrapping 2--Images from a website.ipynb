```python
my_list = list(soup.find_all('loc'))
```

```python
site_maps = [] #empty list
for loc in my_list:
    url = loc.text
    if 'posts' in url:
        print(loc.text)
        site_maps.append(url)
```

```python
site_maps
```

```python
site_map_1 = site_maps[0]
```

```python
response = requests.get(site_map_1)
```

```python
soup = BeautifulSoup(response.text,'xml')
```

```python
#For image
master_list=[]
for loc in soup.find_all('loc'):
    url = loc.text
    master_list.append(url)
```

```python
len(master_list)
```

```python
for image_url in master_list[0:10]:
    print(image_url)
    response = requests.get(image_url)
    soup = BeautifulSoup(response.text, 'html.parser')  # html text

    # Check if the element with class 'download' exists in the HTML
    download_link = soup.find('a', {'class': 'download'})
    if download_link:
        png_url = download_link['href']
        image = requests.get(png_url)  # image name
        image_title = image_url.split('/')[-1] + '-' + png_url.split('/')[-1]

        with open(image_title, 'wb') as file:  # wb-write mode
            file.write(image.content)
    else:
        print("Download link not found for", image_url)
```
